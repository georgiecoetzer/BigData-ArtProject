```python
import pandas as pd
import numpy as np
```

```python
def GBP_to_Dolar(amount):
    dollars = amount * 1.25
    return dollars
```

```python
def EUR_to_Dolar(amount):
    dollars = amount * 1.05
    return dollars
```

```python
def HKD_to_Dolar(amount):
    dollars = amount * 0.13
    return dollars
```

```python
def CHF_to_Dolar(amount):
    dollars = amount * 1.02
    return dollars
```

```python
def CNY_to_Dolar(amount):
    dollars = amount * 0.15
    return dollars
```

## Gets the currency from the estimate and makes it a feature

```python
currency = []
for estimates in ArtInfo["Estimate"]:
    try:
        splitEstimate = estimates.split()
        currency.append(splitEstimate[0])
    except:
        currency.append("0")
        
ArtInfo.insert(loc=5, column='Currency', value=currency)
```

## Converts price from string to float

```python
price_list = []

for price in ArtInfo["price"]:
    try:
        splitPrice = price.split()
        price_list.append(splitPrice[1])
    except:
        price_list.append("0")

float_prices = []   

for price in price_list:
    No_commas = price.replace(",", "")
    No_commas = float(No_commas)
    float_prices.append(No_commas)
        
ArtInfo["price"] = float_prices
```

## splits between low and high estimates

```python
low_high = []
low_estimate = []
high_estimate = []


for a,b in zip(ArtInfo["Currency"], ArtInfo["Estimate"]): 
    b = b.replace(",", "")
    b = b.replace(a, "")
    b = b.replace("–", "")
    b = b.replace("-", "") 
    b = b.replace("   ", " ")
    
    try:
        low_estimate.append(float(b.split()[0]))
    except:
        low_estimate.append(0)
        
    try:
        high_estimate.append(float(b.split()[1]))
    except:
        high_estimate.append(0)       
        
ArtInfo.insert(loc=8, column='low_estimate', value=low_estimate)
ArtInfo.insert(loc=9, column='high_estimate', value=high_estimate)
ArtInfo.drop('Estimate', inplace=True, axis=1)
```

## currency Conversion using 2022 values (Not perfect)

```python
price_usd = []
low_est_usd = []
high_est_usd = []
currency = []

for cur, price, low_est, high_est in zip(ArtInfo["Currency"], ArtInfo["price"], ArtInfo["low_estimate"], ArtInfo["high_estimate"]):
    if cur == "USD":
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    elif cur == "GBP":
        price = GBP_to_Dolar(price)
        low_est = GBP_to_Dolar(low_est)
        high_est = GBP_to_Dolar(high_est)
        
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    elif cur == "EUR":
        price = EUR_to_Dolar(price)
        low_est = EUR_to_Dolar(low_est)
        high_est = EUR_to_Dolar(high_est)
        
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    elif cur == "HKD":
        price = HKD_to_Dolar(price)
        low_est = HKD_to_Dolar(low_est)
        high_est = HKD_to_Dolar(high_est)
        
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    elif cur == "CNY":
        price = CNY_to_Dolar(price)
        low_est = CNY_to_Dolar(low_est)
        high_est = CNY_to_Dolar(high_est) 
        
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    elif cur == "CHF":
        price = CHF_to_Dolar(price)
        low_est = CHF_to_Dolar(low_est)
        high_est = CHF_to_Dolar(high_est)
        
        currency.append("USD")
        price_usd.append(round(price, 2))
        low_est_usd.append(round(low_est, 2))
        high_est_usd.append(round(high_est, 2))
    else:
        currency.append("Error")
        price_usd.append(0)
        low_est_usd.append(0)
        high_est_usd.append(0)
        

ArtInfo["price"] = price_usd
ArtInfo["low_estimate"] = low_est_usd
ArtInfo["high_estimate"] = high_est_usd
ArtInfo["Currency"] = currency
```

## Over_Under_or On Estimate

```python
over_under_est = []

# 1 = over estimate, -1 = under estimate, 0 = in range
for price, low_est, high_est in zip(ArtInfo['price'], ArtInfo['low_estimate'], ArtInfo['high_estimate']):
    if price >= high_est:
        over_under_est.append(1)
    elif price <= low_est:
        over_under_est.append(-1)
    else:
        over_under_est.append(0)
    
ArtInfo.insert(loc=9, column='Est Accuracy', value=over_under_est)
```

## inconsistancy cleaning: (Not nesesarily bad data but not 100% cleanable)

```python
birth = []
death = []

for b_year, d_year in zip(ArtInfo['Birth'], ArtInfo['Death']):
    try:
        birth.append(int(b_year))
    except:
        birth.append(0)
    try:
        death.append(int(d_year))
    except:
        death.append(-1)

ArtInfo["Birth"] = birth
ArtInfo["Death"] = death
```

```python
ArtInfo.drop(ArtInfo.index[ArtInfo['Birth'] == 0], inplace=True)
ArtInfo.drop(ArtInfo.index[ArtInfo['Birth'] <= 1500], inplace=True)
ArtInfo.drop(ArtInfo.index[ArtInfo['Death'] == -1], inplace=True)
ArtInfo.drop(ArtInfo.index[ArtInfo['Currency'] != "USD"], inplace=True)
```

## Artist Age

```python
Age = []
Alive = []
ADS = []

for b_year, d_year, s_year in zip(ArtInfo['Birth'], ArtInfo['Death'], ArtInfo['close date']):
    # If artist is not yet dead
    if d_year == 0:
        Current_Age = 2022 - b_year
        Alive.append(1)
        Age.append(Current_Age)
    # if artist is dead
    else:
        Age_at_death = d_year - b_year
        Alive.append(0)
        Age.append(Age_at_death)
        
            
ArtInfo.insert(loc=3, column='Age', value=Age)
ArtInfo.insert(loc=4, column='Alive?', value=Alive)
ArtInfo.drop(ArtInfo.index[ArtInfo['Age'] > 100], inplace=True)
ArtInfo.drop(ArtInfo.index[ArtInfo['Age'] < 10], inplace=True)
```

## Gathers Artist Popularity

```python
from bs4 import BeautifulSoup
import requests, lxml
import bs4
import re

headers = {
    'User-agent':
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.102 Safari/537.36 Edge/18.19582"
}

def getPopularity(search_term):
    "Gives you the number of hits on yahoo"
    html = requests.get('https://search.yahoo.com/search?p={0}'.format(search_term),headers=headers, ).text
    soup = BeautifulSoup(html, 'lxml')
    results = soup.body.find_all(string=re.compile('.*{0}.*'.format('search results')), recursive=True)
    return results[0][6:-15]
```

```python
Artist_popularity = []

for items in ArtInfo["Artist"]:
    try:
        Artist_popularity.append(getPopularity(items))
    except:
        Artist_popularity.append(0)
#     print("Artist:", items, "searches", getPopularity(items))
```

```python
#Adds the popularity column and gets rid of anyone with a popularity of 0
ArtInfo.insert(loc=5, column='popularity', value=Artist_popularity)
ArtInfo.drop(ArtInfo.index[ArtInfo['popularity'] == 0], inplace=True)
```
