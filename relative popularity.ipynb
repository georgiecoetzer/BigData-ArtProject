```python
def relativePopularityAverage(artist_name, artist_df):
    "Returns the average popularity of a given artist"
    
    print(artist_df)
    
    totalSum = 0
    counter = 0

    for each in artist_df[artist_name]:
        totalSum += each
        counter += 1
    
    average = totalSum/counter
    
    return average
```

```python
#Complex version considering diff dates for each painting

def relativePopularity(dataset):
    "Takes in a dataframe with artists and their painting's selling year"
    "Finds the relative popularity of a list of artists using multiple other functions created above."
    
    #make a new dataframe
    artDF = pd.DataFrame() 
    
    counter = 0
    
    #loop
    for rowIndex,artistInfo in dataset.iterrows(): #go through given dataset
        name = artistInfo['Artist'] #find name
        year = artistInfo['closeYear'] #find year
        month = artistInfo['closeMonth'] #find month
        day = artistInfo['close date'] #find day
        
        #make time-sensitive popularity dataframe based on closing date
        artDF = relativePopularityDF(name, year, month, day)
        
        if artDF.empty:
            dataset.at[counter, 'relPop'] = 0
        
        else:
            try:
                dataset.at[counter, 'relPop'] = relativePopularityAverage(name, artDF)

            except KeyError:
                print("relativePopartDF down bad")
                dataset.at[counter, 'relPop'] = 0
        
        #make time-sensitive popularity dataframe for a time one year before the closing date
        artDFEarly = relativePopularityDF(name, year-1, month, day)
        
        if artDFEarly.empty:
            dataset.at[counter, 'relPopEarly'] = 0
        
        else: 
            try:
                dataset.at[counter, 'relPopEarly'] = relativePopularityAverage(name, artDFEarly)

            except KeyError:
                print("relativePopartDFEarly down bad")
                dataset.at[counter, 'relPopEarly'] = 0
            
        counter += 1
        print(dataset.at[counter-1, 'relPop'])
    
    return dataset
```
